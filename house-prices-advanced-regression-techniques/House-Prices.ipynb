# 房价预测
项目地址：https://www.kaggle.com/competitions/house-prices-advanced-regression-techniques

数据特征如下：
* SalePrice - 以美元为单位的房产销售价格。这是您尝试预测的目标变量。
* MSSubClass : 建筑的等级
* MSZoning：分区分类
* LotFrontage : 与房产相连的街道的直线英尺
* LotArea：以平方英尺为单位的地块大小
* Street: 街道类型
* Alley：巷子类型
* LotShape : 房子整体形状
* LandContour : 平整度级别
* Utilities：可用的公用设施类型
* LotConfig : 房屋配置
* LandSlope : 倾斜度
* Neighborhood: 艾姆斯市区范围内的物理位置
* Condition1：靠近主要道路或铁路便利程度
* Condition2：靠近主要道路或铁路便利程度（如果有第二个）
* BldgType : 住宅类型
* HouseStyle : 住宅风格
* OverallQual：整体材料和表面质量
* OverallCond：整体状况评级
* YearBuilt : 原始建造日期
* YearRemodAdd : 改造日期
* RoofStyle : 屋顶类型
* RoofMatl : 屋顶材料
* Exterior1st : 房屋外墙覆盖物
* Exterior2nd : 房屋外墙覆盖物（如果材料不止一种）
* MasVnrType : 砌体饰面类型
* MasVnrArea : 砌体饰面面积（平方英尺）
* ExterQual : 外部材料质量
* ExterCond : 外部材料的现状
* Foundation: 地基类型
* BsmtQual : 地下室的高度
* BsmtCond : 地下室的一般状况
* BsmtExposure : 花园级地下室墙壁
* BsmtFinType1 : 地下室装饰质量
* BsmtFinSF1 : 地下室装饰面积
* BsmtFinType2：地下室装饰质量（如果存在）
* BsmtFinSF2 : 地下室装饰面积（平方英尺）
* BsmtUnfSF : 未装饰的地下室面积（平方英尺）
* TotalBsmtSF : 地下室总面积（平方英尺）
* Heating：供暖类型
* HeatingQC : 供暖质量和条件
* CentralAir : 中央空调
* Electrical：电气系统
* 1stFlrSF : 一楼面积
* 2ndFlrSF : 二楼面积
* LowQualFinSF : 低质量装饰面积（平方英尺）
* GrLivArea：地上（地面）居住面积（平方英尺）
* BsmtFullBath : 地下室全浴室
* BsmtHalfBath : 地下室半浴室
* FullBath : 高档全浴室
* HalfBath : 高档半浴室
* Bedroom：地下室以上的卧室数量
* Kitchen: 厨房数量
* KitchenQual : 厨房质量
* TotRmsAbvGrd：地上除卧室以外的房间数
* Functional：房屋功能性评级
* Fireplaces: 壁炉数量
* FireplaceQu : 壁炉质量
* GarageType : 车库位置
* GarageYrBlt : 车库建成年份
* GarageFinish : 车库的内部装修
* GarageCars : 车库车容量大小
* GarageArea : 车库面积（平方英尺）
* GarageQual : 车库质量
* GarageCond : 车库条件
* PavedDrive : 铺好的车道情况
* WoodDeckSF : 木地板面积（平方英尺）
* OpenPorchSF : 开放式门廊区面积（平方英尺）
* EnclosedPorch : 封闭门廊区面积（平方英尺）
* 3SsnPorch : 三个季节门廊区面积（平方英尺）
* ScreenPorch : 屏幕门廊面积（平方英尺）
* PoolArea : 泳池面积（平方英尺）
* PoolQC : 游泳池质量
* Fence：围墙质量
* MiscFeature：其他特征未涵盖的其他功能
* MiscVal : 其他杂项特征值
* MoSold : 卖出月份
* YrSold : 卖出年份
* SaleType : 交易类型
* SaleCondition : 交易条件

```python
import warnings
warnings.filterwarnings("ignore")
# 离群点处理
# OverallQual的离群点
train_data.drop((train_data[(train_data['OverallQual'] == 4) & (train_data['SalePrice'] > 200000)].index)|(train_data[(train_data['OverallQual'] == 8) & (train_data['SalePrice'] > 500000)].index)|(train_data[(train_data['OverallQual'] == 10) & (train_data['SalePrice'] > 700000)].index), inplace=True)
# GrLivArea的离群点
train_data.drop(train_data[(train_data['GrLivArea'] > 4000) & (train_data['SalePrice'] < 200000)].index, inplace=True)
# TotalBsmtSF的离群点
train_data.drop(train_data[(train_data['TotalBsmtSF'] > 6000) & (train_data['SalePrice'] < 200000)].index, inplace=True)
# 1stFlrSF的离群点
train_data.drop(train_data[(train_data['1stFlrSF'] > 4000) & (train_data['SalePrice'] < 200000)].index, inplace=True)
# TotRmsAbvGrd的离群点
train_data.drop(train_data[train_data['TotRmsAbvGrd'] == 14].index, inplace=True)
# YearBuilt的离群点
train_data.drop((train_data[(train_data['YearBuilt'] < 1900) & (train_data['SalePrice'] > 400000)].index) | (train_data[(train_data['YearBuilt'] > 1980) & (train_data['SalePrice'] > 700000)].index), inplace=True)
# YearRemodAdd的离群点
train_data.drop(train_data[(train_data['YearRemodAdd'] > 1900) & (train_data['SalePrice'] > 700000)].index, inplace=True)
```

```python
# 合并数据集，方便同时对两个数据集进行清洗
full = pd.concat([train_data, test_data], axis=0, ignore_index=True)
# 删除部分影响小的变量
full.drop(['Utilities', 'RoofMatl', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'Heating', 'LowQualFinSF',
           'BsmtFullBath', 'BsmtHalfBath', 'Functional', 'GarageYrBlt', 'GarageArea', 'GarageCond', 'WoodDeckSF',
           'OpenPorchSF', 'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'PoolQC', 'Fence', 'MiscFeature', 'MiscVal'],
           axis=1, inplace=True)
full.shape
```

```python
full.info()
```

```python
full.describe()
```

```python
# 缺失值含义分析与删除
count = full.isnull().sum().sort_values(ascending=False)
ratio = count/len(full)
nulldata = pd.concat([count, ratio], axis=1, keys=['count', 'ratio'])
nulldata
```

```python
nulldata = nulldata[nulldata['ratio'] > 0]
# 删除缺失多的值(Alley)
full.drop(nulldata[nulldata['ratio'] > 0.8].index, axis=1, inplace=True)
```

```python
# 处理其他缺失值时需要先对变量取值进行分许，如FireplaceQu等，并不是缺失，而是表示没有壁炉.
cols = ['FireplaceQu', 'GarageQual', 'GarageFinish', 'GarageType', 'BsmtExposure', 'BsmtCond', 'BsmtQual', 'BsmtFinType1', 'BsmtFinType2']
for col in cols:
    full[col] = full[col].fillna('None')
```

```python
# 处理含有缺失值的属性
cols2 = ['MasVnrType', 'MSZoning', 'KitchenQual', 'Electrical', 'SaleType', 'Exterior1st', 'Exterior2nd']
for col in cols2:
    full[col] = full[col].fillna(full[col].mode()[0])
```

```python
# 处理含有缺失值的数值型属性
# TotalBsmtSF和GarageCars缺失是因为没有
cols3 = ['TotalBsmtSF', 'GarageCars']
for col in cols3:
    full[col] = full[col].fillna(0)
```

```python
# 还剩下LotFrontage数值型，可以用平均值填充
full['LotFrontage'] = full['LotFrontage'].fillna(full['LotFrontage'].mean())
```

```python
full.info()
```

```python
# 将年份及类别的数值型转换为字符串
full['MSSubClass'] = full['MSSubClass'].astype(str)
full['YrSold'] = full['YrSold'].astype(str)
full['MoSold'] = full['MoSold'].astype(str)
full['OverallCond'] = full['OverallCond'].astype(str)
full['YearBuilt'] = full['YearBuilt'].astype(str)
full['YearRemodAdd'] = full['YearRemodAdd'].astype(str)
```

```python
# 顺序特征编码
def map_values(df):
    df['MSZoning'] = df['MSZoning'].map({'C (all)': 1, 'RM': 2, 'RH': 2, 'RL': 3, 'FV': 4})
    df['Neighborhood'] = df['Neighborhood'].map({'MeadowV': 1,
                                               'IDOTRR': 2, 'BrDale': 2,
                                               'OldTown': 3, 'Edwards': 3, 'BrkSide': 3,
                                               'Sawyer': 4, 'Blueste': 4, 'SWISU': 4, 'NAmes': 4,
                                               'NPkVill': 5, 'Mitchel': 5,
                                               'SawyerW': 6, 'Gilbert': 6, 'NWAmes': 6,
                                               'Blmngtn': 7, 'CollgCr': 7, 'ClearCr': 7, 'Crawfor': 7,
                                               'Veenker': 8, 'Somerst': 8, 'Timber': 8,
                                               'StoneBr': 9,
                                               'NoRidge': 10, 'NridgHt': 10})

    df['HouseStyle'] = df['HouseStyle'].map({'1.5Unf':1, '1Story':1,
                                           '1.5Fin':2, '2.5Unf':2, 'SFoyer':2, '2Story':2,
                                            'SLvl':3, '2.5Fin':3})

    df['MasVnrType'] = df['MasVnrType'].map({'None':1, 'BrkCmn':2, 'BrkFace':3, 'Stone':4})

    df['ExterQual'] = df['ExterQual'].map({'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex':5})

    df['ExterCond'] = df['ExterCond'].map({'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex':5})

    df['Foundation'] = df['Foundation'].map({'Slab':1, 'BrkTil':2, 'CBlock':2, 'Stone':2, 'Wood':3, 'PConc':4})

    df['BsmtQual'] = df['BsmtQual'].map({'None':1, 'Po':2, 'Fa':3, 'TA':4, 'Gd':5, 'Ex':6})

    df['BsmtCond'] = df['BsmtCond'].map({'None':1, 'Po':2, 'Fa':3, 'TA':4, 'Gd':5, 'Ex':6})

    df['BsmtExposure'] = df['BsmtExposure'].map({'None':1, 'No':2, 'Mn':3, 'Av':4, 'Gd':5})

    df['BsmtFinType1'] = df['BsmtFinType1'].map({'None':1, 'Unf':2, 'LwQ':3, 'Rec':4, 'BLQ':5, 'ALQ':6, 'GLQ':7})

    df['BsmtFinType2'] = df['BsmtFinType2'].map({'None':1, 'Unf':2, 'LwQ':3, 'Rec':4, 'BLQ':5, 'ALQ':6, 'GLQ':7})

    df['HeatingQC'] = df['HeatingQC'].map({'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex':5})

    df['KitchenQual'] = df['KitchenQual'].map({'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex':5})

    df['FireplaceQu'] = df['FireplaceQu'].map({'None':1, 'Po':2, 'Fa':3, 'TA':4, 'Gd':5, 'Ex':6})

    df['GarageFinish'] = df['GarageFinish'].map({'None':1, 'Unf':2, 'RFn':3, 'Fin':4})

    df['GarageQual'] = df['GarageQual'].map({'None':1, 'Po':2, 'Fa':3, 'TA':4, 'Gd':5, 'Ex':6})

    df['PavedDrive'] = df['PavedDrive'].map({'N':1, 'P':2, 'Y':3})
map_values(full)
```

```python
# 添加一个关于总面积的特征
full['TotalSF'] = full['TotalBsmtSF'] + full['1stFlrSF'] + full['2ndFlrSF']
```

```python
# 时间特征编码
from sklearn import preprocessing 
lab = preprocessing.LabelEncoder()
full['YearBuilt'] = lab.fit_transform(full['YearBuilt'])
full['YearRemodAdd'] = lab.fit_transform(full['YearRemodAdd'])
full['YrSold'] = lab.fit_transform(full['YrSold'])
```

```python
import numpy as np
# 数据做一备份
full_fe = full.copy()
full_fe.drop(['SalePrice', 'Id'], axis=1, inplace=True)
# 对于偏度大于1的定量变量进行对数变换
full_numeric = full_fe.select_dtypes(exclude='object')
skew = full_numeric.apply(lambda x: x.skew())
skew_features = skew[abs(skew) >= 1].index
full_fe[skew_features] = np.log1p(full_fe[skew_features])
```

```python
# one-hot编码
full_fe = pd.get_dummies(full_fe)
```

```python
# 将数值变量标准化
numerical_col = full_fe.columns[full_fe.dtypes!='object']
means = full_fe.loc[:,numerical_col].mean()
std = full_fe.loc[:,numerical_col].std()
full_fe.loc[:,numerical_col] = (full_fe.loc[:,numerical_col] - means)/std
```

```python
# 建立训练数据集和测试数据集
from sklearn.model_selection import train_test_split
source_X = full_fe.loc[0:1451, :]
source_y = full.loc[0:1451, 'SalePrice']
pred_X = full_fe.loc[1452:, :]
train_X, test_X, train_y, test_y = train_test_split(source_X, source_y, train_size=0.8, random_state=0)
print('原始数据集特征：', source_X.shape, '训练数据集特征：', train_X.shape, '测试数据集特征：', test_X.shape)
print('原始数据集标签：', source_y.shape, '训练数据集标签：', train_y.shape, '测试数据集标签：', test_y.shape)
```

```python
# 随机森林
from sklearn.ensemble import RandomForestRegressor
clf = RandomForestRegressor(random_state=0)
clf = clf.fit(train_X, train_y)
clf.score(test_X,test_y)
```

```python
# 使用模型对预测数据集中的生存情况进行预测
pred_y = clf.predict(pred_X)
id = full.loc[1452:, 'Id']
pred_df = pd.DataFrame({
    'Id': id,
    'SalePrice': pred_y
})
pred_df.head()
```
